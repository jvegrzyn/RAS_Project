```python
import pandas as pd

h5_file_path = 'Sandusky unsteady fluegg data.h5'
hydraulic_data = pd.read_hdf(h5_file_path, 'unsteady')
```

```python
hydraulic_data.head()
```

```python
hydraulic_data.to_csv(r'hydraulic_data.csv')
```

```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
cumulative_distance = hydraulic_data['CumlDistance_km'].unique()
x_len = cumulative_distance.shape[0]
```

```python
datetime_stamps = hydraulic_data.index.get_level_values(0).unique()
time_since_start = (datetime_stamps - datetime_stamps[0]).total_seconds()/(60*60)
t_len = time_since_start.shape[0]
```

```python
velocity = np.reshape(hydraulic_data['Vmag_mps'].values, (t_len, x_len))
z = velocity
X, Y = np.meshgrid(cumulative_distance, time_since_start)
```

```python
plt.contourf(X, Y, z, 100, cmap='RdGy') #amplitude 0.5 m, Period 10 h
plt.colorbar(label = 'Velocity (m/s)');
plt.xlabel('River Distance (km)')
plt.ylabel('Time (h)')
```

```python
plt.contourf(X, Y, z, 100, cmap='RdGy') #Phase Lag
plt.colorbar(label = 'Velocity (m/s)');
plt.xlabel('River Distance (km)')
plt.ylabel('Time (h)')
```
