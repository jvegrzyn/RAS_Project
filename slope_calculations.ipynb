```python
import numpy as np
import pandas as pd
```

```python
#Before loading in insert column of datetimenum in excel.
df = pd.read_csv('hydraulic_data.csv')
```

```python
#Create array of 32 whole number kilometers.
count = 1
kilometers = []
for i in range(32):
    kilometers.append(count)
    count += 1
kilometers = np.array(kilometers)
```

```python
distances = np.array(df['CumlDistance_km'])
```

```python
#Create array of nearest kilometers to whole number kilometers in python-ras data.
nearest_km = []
for i in range(len(kilometers) + 1):
    a = distances.flat[np.abs(distances - i).argmin()]
    nearest_km.append(a)
nearest_km = np.array(nearest_km)
nearest_km
```

```python
df_wholekm = df[df['CumlDistance_km'].isin(nearest_km)]
depths = np.array(df_wholekm['Depth_m'])
distances = np.array(df_wholekm['CumlDistance_km'])
discharge = np.array(df_wholekm['Q_cms'])
```

```python
df_wholekm.head()
```

```python
#Calculate the slope over the kilometers.
depth_slope = []
dis_slope = []
for i in range(len(df_wholekm)):
    if i >= 1:
        s = (depths[i] - depths[i - 1]) / (distances[i] - distances[i - 1]) / 1000
        depth_slope.append(s)
        
        q = (discharge[i] - discharge[i - 1]) / (distances[i] - distances[i - 1])
        dis_slope.append(q)
        
depth_slope = np.array(depth_slope)
depth_slope = np.insert(depth_slope, 0, 0)
dis_slope = np.array(dis_slope)
dis_slope = np.insert(dis_slope, 0, 0)
```

```python
df_wholekm.insert(10, 'depth_slope (m/m)', depth_slope)
df_wholekm.insert(11, 'discharge_slope (m2/s)', dis_slope)
```

```python
df_wholekm_distance_sorted = df_wholekm.sort_values('CumlDistance_km', ascending = True, inplace = False)
```
